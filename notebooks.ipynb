```python
import pandas as pd
n = 5

class CollectVars:
    def __init__(self, **kwargs) -> None:
        self.dict = {k:[v] for k,v in kwargs.items()}
    
    def load(self, **kwargs):
        for k,v in kwargs.items():
            if k in self.dict:
                self.dict[k].append(v)
            else:
                self.dict[k] = [v]
            
    def data_frame(self):
        return pd.DataFrame(self.dict)

collect_vars = CollectVars()
for i in range(n):
    for j in range(i + 1, n):
        for k in range(i, j):
            collect_vars.load(**dict(i=i,j=j,k=k))
collect_vars.data_frame()
```

```python
collect_vars = CollectVars()
for i in range(n):
    for j in range(i + 1, n):
        for k in range(i, j):
            collect_vars.load(**dict(i=i,j=j,k=k, s=(k-i)))
collect_vars.data_frame()
```
